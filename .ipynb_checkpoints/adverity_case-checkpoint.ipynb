```python
# Use Dask to process data in parallel
import dask.dataframe as dd
data = dd.read_csv('data/train', dtype={'id': 'float64'})
```

```python
# Define the template for data aggregation by hour and calculate the CTR
# CTR = number of clicks / number of impressions
# - number of clicks is calculated by summing up the click-variable
# - number of impression is calculated by *counting* the click-variable

dask_df = data.groupby(['hour'])['click'].agg(['sum', 'count'])
dask_df['CTR'] = dask_df['sum'] / dask_df['count']
```

```python
# Execute the template
df = dask_df.compute()
```

```python
# Convert `hour` to a datetime for nicer plot, since we just did a groupby using `hour`, it will be in the index
import pandas as pd
df['datetime'] = pd.to_datetime(df.index, format='%y%m%d%H')
```

```python
# Construct the rolling windows for the mean and std calculation
window_length = 12 # The length of the rolling window in number of samples
df['CTR_mean'] = df['CTR'].rolling(window_length).mean()
df['CTR_std'] = df['CTR'].rolling(window_length).std()
```

```python
# Build the outlier detector as specified in the case description
import numpy as np

def outlier_detector(row):
    x = row['CTR']
    mu = row['CTR_mean']
    sigma = row['CTR_std']
    
    if np.abs(x - mu) / sigma > 1.5:
        return 'Outlier'
    else:
        return 'Inlier'
    
df['datapoint_type'] = [outlier_detector(row) for _, row in df.iterrows()]
```

```python
# We're offsetting the data to be plotted by `window_length` samples, since there's only information about the moving avg and moving std for subsequent samples
# Had we not done this, one might have interpreted the results as there are no outliers in the first `window_length` samples
labelled_df = df.iloc[window_length:]
```

```python
# Plot the timeseries, and color the individual datapoints according to the `datapoint_type` field
import seaborn as sns
sns.set(rc={'figure.figsize': (15, 8)})

sns.lineplot(data=labelled_df, x='datetime', y='CTR', palette=['C1', 'C0'])
sns.scatterplot(data=labelled_df, x='datetime', y='CTR', hue='datapoint_type', palette=['C1', 'C0'])
```
